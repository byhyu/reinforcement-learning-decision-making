```python
#import time
import gym
import numpy as np
import random
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import Adam

from collections import deque

class DQN:
    def __init__(self, env):
        self.env     = env
        self.memory  = deque(maxlen=int(1e5))
        
        self.gamma = 0.99
        self.learning_rate = 5e-4 
        self.tau = 1e-3   

        self.model        = self.create_model()
        self.target_model = self.create_model()
        
        self.t_step = 0

    def create_model(self):
        model = Sequential()
        state_shape = self.env.observation_space.shape
        model.add(Dense(64, input_dim=state_shape[0], activation="relu"))
        model.add(Dense(64, activation="relu"))
        model.add(Dense(self.env.action_space.n))
        model.compile(loss="mean_squared_error",optimizer=Adam(lr=self.learning_rate))
        return model

    def act(self, state, eps):
        if np.random.random() < eps:
            return self.env.action_space.sample()
        return np.argmax(self.model.predict(state)[0])

    def remember(self, state, action, reward, new_state, done):
        self.memory.append([state, action, reward, new_state, done])

    def replay(self):
        batch_size = 64
        
        self.t_step = (self.t_step + 1) % 4
        if self.t_step == 0:
            if len(self.memory) < batch_size: 
                return
            states, targets_f = [], []
            samples = random.sample(self.memory, batch_size)
            for state, action, reward, new_state, done in samples:
                if done:
                    target = reward
                else:
                    Q_future =  np.amax(self.target_model.predict(new_state)[0])
                    target = reward + Q_future * self.gamma

                target_f = self.model.predict(state)
                target_f[0][action] = target
                states.append(state[0])
                targets_f.append(target_f[0])
            self.model.fit(np.array(states), np.array(targets_f), epochs=1, verbose=0)
            self.target_train()

    def target_train(self):
        #print("trigger target_train")
        weights = self.model.get_weights()
        target_weights = self.target_model.get_weights()
        for i in range(len(target_weights)):
            target_weights[i] = weights[i] * self.tau + target_weights[i] * (1 - self.tau)
        self.target_model.set_weights(target_weights)

    def save_model(self, fn):
        self.model.save(fn)
```

```python
def main():
    env = gym.make("LunarLander-v2")
    
    epsilon = 1.0
    epsilon_min = 0.01
    epsilon_decay = 0.995

    trials  = 2000
    trial_len = 1000
    
    scores = []                        # list containing scores from each episode
    scores_window = deque(maxlen=100)
    
    # updateTargetNetwork = 1000
    dqn_agent = DQN(env=env)
    steps = []
    for trial in range(trials):
        #print(trial)
        score = 0
        cur_state = env.reset().reshape(1,8)
        #print('\rtrial {}\Epsilon: {:.5f}\n'.format(trial, epsilon), end="")
        for step in range(trial_len):
            env.render()
            #start_time = time.time()
            
            action = dqn_agent.act(cur_state ,epsilon)
            new_state, reward, done, _ = env.step(action)
            score += reward
            #print("--- %s seconds getting reward and next state ---" % (time.time() - start_time))
            
            # reward = reward if not done else -20
            new_state = new_state.reshape(1,8)
            dqn_agent.remember(cur_state, action, reward, new_state, done)
            
            #start_time = time.time()
            dqn_agent.replay()       # internally iterates default (prediction) model
            #dqn_agent.target_train() # iterates target model
            #print("--- %s seconds getting reward and next state ---" % (time.time() - start_time))
            cur_state = new_state
            if done:
                break   
            
        scores_window.append(score)       # save most recent score
        scores.append(score)
        epsilon *= epsilon_decay
        epsilon = max(epsilon_min, epsilon)
        
        print('\rtrial {}\tAverage Score: {:.2f} \t with epsilon{:.2f}'.format(trial, np.mean(scores_window),epsilon), end="")
        if trial % 100 == 0:
            print('\rtrial {}\tAverage Score: {:.2f}'.format(trial, np.mean(scores_window)))
        if np.mean(scores_window)>=200.0:
            print('\nEnvironment solved in {:d} trial!\tAverage Score: {:.2f}'.format(trial-100, np.mean(scores_window)))
            dqn_agent.save_model("success.h5")
            break                     
    env.close()

if __name__ == "__main__":
    main()
```
